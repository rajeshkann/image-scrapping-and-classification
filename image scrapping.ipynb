```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
from bs4 import BeautifulSoup
import requests
```

```python
amazon=webdriver.Chrome("chromedriver.exe")
amazon.get("https://www.amazon.in/")
```

```python
list1=["Trousers (men)","Sarees (women)","Jeans (men)"]
for i in list1:
    select=amazon.find_element_by_id("twotabsearchtextbox")
    select.send_keys(Keys.CONTROL+"a",Keys.DELETE)
    select.send_keys(i,Keys.ENTER)
    image=[]
    for r in range(5):
        images=amazon.find_elements_by_xpath("//div[@class='s-image-padding']//img")
        for w in images:
            image.append(w.get_attribute("src"))
        time.sleep(2)
        amazon.find_element_by_xpath("//*[contains(text(),'Next')]").click()
        time.sleep(2)
    for o in range(len(image)):
        if o>200:
            break
        print("complete {0} in 200".format(o))
        if i=="Sarees (women)":
            file=open("D:\project images\sarees\saree"+str(o)+".jpg","wb")
            response=requests.get(image[o])
            file.write(response.content)
        elif i=="Jeans (men)":
            file=open("D:\project images\jeans\jean"+str(o)+".jpg","wb")
            response=requests.get(image[o])
            file.write(response.content)
        else:
            file=open("D:\project images\trousers\jean"+str(o)+".jpg","wb")
            response=requests.get(image[o])
            file.write(response.content)
```
